```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv')

# pd.Categorical (https://pandas.pydata.org/docs/reference/api/pandas.Categorical.html)
# this converts the 'continent' column in the DataFrame data into a categorical data type using pandas
# similar to organizing your toys into labelled bins 
# ie. say you have a box crayons but only in red, blue, yellow
data['continent'] = pd.Categorical(data['continent'])
data.head()
# why?
# this helps the computer save space (eg. color #1 instead of "red" every time); 
# work faster; 
# know these are categories, not text

fig, ax = plt.subplots(figsize=(10, 10))

# Subset of the data for year 1952
data1952 = data[data.year == 1952]

# Scatterplot
ax.scatter(
    x = data1952['lifeExp'], 
    y = data1952['gdpPercap'], 
    s = data1952['pop']/50000, 
    c = data1952['continent'].cat.codes, 
    cmap = "Accent", 
    alpha = 0.6, 
    edgecolors = "white", 
    linewidth = 2
)
 
# Add titles (main and on axis)
plt.yscale('log')
ax.set_xlabel("Life Expectancy")
ax.set_ylabel("GDP per Capita")
ax.set_title("Year 1952")
ax.set_ylim(100,50000)
ax.set_xlim(30, 75)

plt.show()
```
